```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
purchase_behaviour  = pd.read_csv("/content/purchase_behaviour.csv")
transactions = pd.read_csv("/content/transaction_data.csv")
```

## Data Cleaning and Preparation

```python
# Convert DATE to datetime
transactions['DATE'] = pd.to_datetime(transactions['DATE'], unit='D', origin='1899-12-30')

# Clean product names by removing weights
transactions['PROD_NAME'] = transactions['PROD_NAME'].str.replace(r'\d+[gG]$', '', regex=True).str.strip()

# Merge datasets
merged_data = pd.merge(transactions, purchase_behaviour, on='LYLTY_CARD_NBR')
```

## 1. Identify top 3 most profitable products

```python
product_profitability = merged_data.groupby('PROD_NAME')['TOT_SALES'].sum().sort_values(ascending=False)
top_products = product_profitability.head(3)
```

## 2. Identify characteristics of most loyal customers

```python
# Calculate loyalty metrics (frequency and total spending)
customer_loyalty = merged_data.groupby('LYLTY_CARD_NBR').agg(
    transaction_count=('TXN_ID', 'count'),
    total_spent=('TOT_SALES', 'sum')
).reset_index()

# Get top 10% most loyal customers
top_customers = customer_loyalty.nlargest(int(len(customer_loyalty)*0.1), 'transaction_count')

# Merge with demographic data
customer_characteristics = pd.merge(top_customers, purchase_behaviour, on='LYLTY_CARD_NBR')

# Analyze characteristics
lifestage_dist = customer_characteristics['LIFESTAGE'].value_counts(normalize=True)
premium_dist = customer_characteristics['PREMIUM_CUSTOMER'].value_counts(normalize=True)
```

## 3. Hypothesis testing - Why loyal customers prefer products

```python
# Get top products for loyal customers
loyal_customer_data = merged_data[merged_data['LYLTY_CARD_NBR'].isin(top_customers['LYLTY_CARD_NBR'])]
loyal_top_products = loyal_customer_data.groupby('PROD_NAME')['TOT_SALES'].sum().sort_values(ascending=False).head(5)
```

## Plots

```python
def plot_results():
    plt.figure(figsize=(15, 10))

    # Customize color palette
    colors = ['#1ca089', '#9367ba', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

    # Top products plot
    plt.subplot(2, 2, 1)
    top_products.plot(kind='bar', color=colors[0]) # Apply custom color
    plt.title('Top 3 Most Profitable Products', fontsize=14) # Increase title font size
    plt.ylabel('Total Sales', fontsize=12) # Increase label font size
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

    # Customer characteristics plot
    plt.subplot(2, 2, 2)
    lifestage_dist.plot(kind='pie', autopct='%1.1f%%', colors=colors[1:4]) # Apply custom color palette
    plt.title('Lifestage Distribution of Loyal Customers', fontsize=14)

    plt.subplot(2, 2, 3)
    premium_dist.plot(kind='pie', autopct='%1.1f%%', colors=colors[4:6]) # Apply custom color palette
    plt.title('Premium Customer Distribution', fontsize=14)

    plt.subplot(2, 2, 4)
    loyal_top_products.head(3).plot(kind='bar', color=colors[6]) # Apply custom color
    plt.title('Top Products Preferred by Loyal Customers', fontsize=14)
    plt.ylabel('Total Sales', fontsize=12)
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.savefig('marketing_insights.png')
    plt.show()
```

```python
# Generate report
def generate_report():
    report = f"""
    Marketing Insights Report
    =========================

    1. Top 3 Most Profitable Products:
    {top_products.to_string()}

    2. Loyal Customer Characteristics:
    - Life Stage Distribution:
    {lifestage_dist.to_string()}

    - Premium Customer Distribution:
    {premium_dist.to_string()}

    3. Hypothesis for Customer Preference:
    Loyal customers primarily purchase {loyal_top_products.index[0]} and similar products.
    This preference might be driven by:
    - Strong brand loyalty to particular snack brands
    - Regular consumption patterns for these specific products
    - Price sensitivity matching their customer segment
    - Product availability in frequently visited stores
    """
    with open('marketing_report.txt', 'w') as f:
        f.write(report)
    return report

# Execute analysis
plot_results()
print(generate_report())
```
